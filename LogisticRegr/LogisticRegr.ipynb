```python
import kagglehub
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
from sklearn.model_selection import train_test_split
from sklearn.preprocessing import MinMaxScaler,StandardScaler
from sklearn.metrics import confusion_matrix, classification_report
```

```python
df=pd.read_csv("BankNote_Authentication.csv")
df.head()
```

```python
X = df.drop(columns="class")
y = df["class"]

X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.3, random_state=42, shuffle=True)
```

```python
min_max_Scaler = MinMaxScaler()

X_train_scaled = min_max_Scaler.fit_transform(X_train)
X_test_scaled = min_max_Scaler.transform(X_test)
```

```python
X_train_scaled
```

```python
class LogisticRegression:
    def __init__(self, learning_rate=0.01,batch=10,epochs=1000):
        self.lr = learning_rate
        self.bias = 0
        self.slope = None
        self.batch=batch
        self.epochs=epochs
        
        
    def fit(self, X, y, iter=1000):
        X = np.array(X)
        y = np.array(y).flatten()
        
        if X.ndim == 1:
            X = X.reshape(-1, 1)
        
        self.slope = np.zeros(X.shape[1])
        n = len(X)
            
        for _ in range(self.epochs):
            X=X.sample(frac=1)
            for _ in range(self.batch):
                z = np.dot(X, self.slope) + self.bias
                sigmoid_pred = self.sigmoid(z)
                
                error = sigmoid_pred - y
                
                dw = np.dot(X.T, error) / n
                db = np.sum(error) / n
            
                self.slope -= self.lr * dw
                self.bias -= self.lr * db
        
    def predict(self, X):
        X = np.array(X)
        z = np.dot(X, self.slope) + self.bias
        probabilities = self.sigmoid(z)
        return np.where(probabilities >= 0.5, 1, 0)
        
    def sigmoid(self, z):
        return 1 / (1 + np.exp(-z))
```

```python
model=LogisticRegression(learning_rate=0.1)
model.fit(X_train_scaled,y_train,iter=10000)
results=model.predict(X_test_scaled)
```

```python
print("Confusion Matrix:")
print(confusion_matrix(y_test, results))
```

```python
print("\nClassification Report:")
print(classification_report(y_test, results))
```
